```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
page_view_events_100 = pd.read_csv('/mnt/c/Users/Jovan Bogoevski/StreamsSongs/tables/page_view_events_100.csv')
page_view_events_100.head(5)
```

```python
'''
The attributes of the page_view_events_100.csv dataset are as follows:

ts: Timestamp of the event in milliseconds (integer).
sessionId: Unique identifier for the session (integer).
page: Page viewed by the user (string).
auth: Authorization status of the user (string).
method: HTTP method used (e.g., "GET", "POST") (string).
status: HTTP status code for the page view (integer).
level: Subscription level of the user (e.g., "paid" or "free") (string).
itemInSession: Position of the item in the session (integer).
city: City of the user's location (string).
zip: ZIP code of the user's location (integer).
state: State of the user's location (string).
userAgent: User's browser and OS details (string).
lon: Longitude of the user's location (float).
lat: Latitude of the user's location (float).
userId: Unique identifier for the user (float).
lastName: Last name of the user (string).
firstName: First name of the user (string).
gender: Gender of the user (string).
registration: User's registration time in milliseconds (float).
artist: Artist related to the page view, if applicable (string).
song: Song title related to the page view, if applicable (string).
duration: Duration of the song in seconds, if applicable (float)
'''
```

```python
## Plot 1: Distribution of Page Views by Page Type
plt.figure(figsize=(10, 6))
page_view_events_100['page'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Distribution of Page Views by Page Type")
plt.xlabel("Page Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 3: Page View Counts by User Level
plt.figure(figsize=(8, 5))
page_view_events_100['level'].value_counts().plot(kind='bar', color=['salmon', 'lightgreen'])
plt.title("Page View Counts by User Level")
plt.xlabel("User Level")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

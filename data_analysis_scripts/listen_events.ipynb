```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
listen_events = pd.read_csv('/mnt/c/Users/Jovan Bogoevski/StreamsSongs/tables/listen_events_100.csv')
listen_events.head(5)
```

```python
'''
The attributes of the listen_events_100.csv dataset are:

artist: Name of the artist (string)
song: Title of the song (string)
duration: Duration of the song in seconds (float)
ts: Timestamp of the event in milliseconds (integer)
sessionId: Unique identifier for the session (integer)
auth: Authorization status of the user (string)
level: Subscription level of the user (e.g., "paid" or "free") (string)
itemInSession: Position of the item in the session (integer)
city: City of the user's location (string)
zip: ZIP code of the user's location (integer)
state: State of the user's location (string)
userAgent: User's browser and OS details (string)
lon: Longitude of the user's location (float)
lat: Latitude of the user's location (float)
userId: Unique identifier for the user (integer)
lastName: Last name of the user (string)
firstName: First name of the user (string)
gender: Gender of the user (string)
registration: User's registration time in milliseconds (integer) 
'''
```

```python
'''
4. Geographic Distribution (by City or State)
A bar plot of events by city or state to understand user concentration by region.
'''
```

```python
# Top 10 cities by event count
listen_events['city'].value_counts().head(10).plot(kind='bar', color='teal')
plt.title("Top 10 Cities by Event Count")
plt.xlabel("City")
plt.ylabel("Event Count")
plt.show()
```

```python
#A pie chart or bar plot to show the distribution of users by gender.
```

```python
# Plot gender distribution
listen_events['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightblue'])
plt.title("Gender Distribution")
plt.ylabel("")  # Hide y-axis label for pie chart
plt.show()
```
